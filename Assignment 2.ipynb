### Assign NAs to a file

```python
missing_columns = [col for col in players_22.columns if players_22[col].isnull().sum() > len(players_22) / 2]
```

```python
missing_columns
```

```python
players_22
```

```python
# write to file
na_count = 0
with open ('fifa missing values.txt', 'w') as f:
    for col in players_22.columns:
        missing = players_22[col].isnull().sum()
        if missing > len(players_22)/2:
            na_count += 1
            f.write(f"{col}: {missing}\n")
    f.write(f"Total number of columns to be removed: {na_count}")
```

### Remove them through automation

```python
# read from file
with open('fifa missing values.txt', 'r') as file:
    for line in file:
        drop_miss = line.strip().split(':')[0]  # Read each line, strip whitespace
        if drop_miss in players_22.columns:  # Check if the column exists in the DataFrame
            players_22.drop(drop_miss, axis=1, inplace=True)
```
