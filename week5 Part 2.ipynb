### Assign NAs list to file

```python
with open ('missing values.txt', 'w') as f:
    for col in df.columns:
        missing = df[col].isnull().sum()
        if missing > 245:
            f.write(f"{col}: {missing}\n")
```

### Use file to drop NAs - automated

```python
with open('missing values.txt', 'r') as file:
    for line in file:
        drop_miss = line.strip().split(':')[0]  # Read each line, strip whitespace
        if drop_miss in df.columns:  # Check if the column exists in the DataFrame
            df.drop(drop_miss, axis=1, inplace=True)
```

### Transforming the data - pipelines

```python
from sklearn.impute import SimpleImputer
from numpy import int64, float64
from sklearn.pipeline import Pipeline
from sklearn. compose import ColumnTransformer, make_column_selector
```

```python
cat_pipeline = Pipeline([
        ("impute", SimpleImputer(strategy="most_frequent")),
        # ("standardize", StandardScaler()),
])
```

```python
quant_pipeline = Pipeline([
        ("impute", SimpleImputer(strategy="median")),
        # ("standardize", StandardScaler()),
])
```

```python
full_pipeline = ColumnTransformer([
    ('quant', quant_pipeline, make_column_selector(dtype_include=[float64, int64])),    #int, float because quantitave 
    ('cat', cat_pipeline, make_column_selector(dtype_include=object)) #object because categorical values are objects 
])
```

```python
piped_data = full_pipeline.fit_transform(df)
df = pd.DataFrame(data=piped_data, columns=full_pipeline.get_feature_names_out)
```
